# Introduction

This project is about analyzing economic data of different countries with analyzing following parameters:
- use of ELectricity per Person
- Income per Capita(PPP)
- Gross Domestic Product(GDP)

```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Data Wragling 

```python
# Loading dataset to analyze

df_gdp = pd.read_csv("total_gdp_ppp_inflation_adjusted.csv")

df_electricity = pd.read_csv("electricity_use_per_person.csv")

df_income = pd.read_csv("income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
```

Data doesn't look very appeling for analysis that's why I will reshape every dataset.

```python
# Reshaping dataset by using pd.melt()
df_gdp = pd.melt(df_gdp, id_vars='country', var_name='year', value_name='gdp')
df_income = pd.melt(df_income, id_vars='country', var_name='year', value_name='income')
df_electricity = pd.melt(df_electricity, id_vars='country', var_name='year', value_name='electricity_consumption')
```

Lokking at the reshaped dataset

```python
df_gdp.head()
```

```python
df_income.head()
```

```python
df_electricity.head()
```

Now I want to merge all three dataset in one dataframe.

```python
# Merging the datasets
from functools import reduce
frames = [df_gdp,df_income,df_electricity]
df_merged = reduce(lambda left , right: pd.merge(left,right, on = ['country','year'], how = 'outer'),frames)
```

Look at the merged datasets

```python
df_merged.head()
```

```python
# Sorting the merged dataframe by columns : country and year
df_merged.sort_values(by = ['country','year'], ascending=(True, True),inplace = True)
```

```python
df_merged
```

Looking for the missing values in each column of the dataset

```python
df_merged.isnull().sum()
```

GDP, Income and Electricity_consumption all three of the columns have null values there can be a lot of reasons for the missing data. 

```python
# making a copy of the original dataset before cleaning the missing values
df_merged1 = df_merged.copy()
```

```python
df_merged1.head()
```

Dropping na values from the dataset

```python
df_merged1 = df_merged1.dropna(axis = 0, how='any')
```

```python
df_merged1.head()
```

```python
df_merged1 = df_merged1[df_merged1['year'] >= '1965']
```
